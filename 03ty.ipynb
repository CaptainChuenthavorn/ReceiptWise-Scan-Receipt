```python
import easyocr
from PIL import Image
import cv2
import pytesseract
pytesseract.pytesseract.tesseract_cmd=r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Load the image
# img=cv2.imread(r"temp\4_contour\kbank\4_3_thresh_kbank01.png")
image = Image.open(r"temp\bw_image.jpg")

# Initialize EasyOCR with the desired language model
reader = easyocr.Reader(['th','en'],gpu=False)

# Perform text detection and get the list of detected boxes and texts
results = reader.readtext(image)

# Iterate through the detected boxes and texts
for (bbox, text, prob) in results:
    # Extract the coordinates of the detected box
    x1, y1 = bbox[0]
    x2, y2 = bbox[2]

    # Crop the region from the original image
    cropped_image = image.crop((x1, y1, x2, y2))

    # Use pytesseract to extract text from the cropped image
    cropped_text = pytesseract.image_to_string(cropped_image, lang='tha+eng')

    # Print the extracted text and its bounding box coordinates
    print("Prob:",prob)
    print("Detected Text:", text)
    print("Coordinates:", bbox)
    print("Extracted Text:", cropped_text)
    print("---")
```

```python

import numpy as np

import matplotlib.pyplot as plt
import cv2

kernel = np.ones((2,2),np.uint8)

# load image

img = cv2.imread("E-receipt-Dataset\scb\IMG_1063.JPG")

# Convert BGR to HSV

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# define range of black color in HSV

lower_val = np.array([0,0,0])

upper_val = np.array([179,100,130])

# Threshold the HSV image to get only black colors

mask = cv2.inRange(hsv, lower_val, upper_val)

# Bitwise-AND mask and original image

res = cv2.bitwise_and(img,img, mask= mask)

# invert the mask to get black letters on white background

res2 = cv2.bitwise_not(mask)

# display image

cv2.imshow("img", res)
```
